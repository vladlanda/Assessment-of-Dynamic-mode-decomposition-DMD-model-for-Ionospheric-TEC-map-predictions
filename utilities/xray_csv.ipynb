```python
import pandas as pd
import os
import glob
import re
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# % matplotlib qt
```

```python
euvs_dir = os.path.join(os.path.abspath(os.curdir),'..','EUVs')
years = ['2013','2014']

li = []

def parseCSV2pandas(csv_file):
    n_rows2skip = 138
    pd_frame = pd.read_csv(csv_file,error_bad_lines=False,skiprows=n_rows2skip)
    n_columns = len(list(pd_frame.columns.values))
    if n_columns < 7:
        pd_frame = pd.read_csv(csv_file,error_bad_lines=False,skiprows=n_rows2skip+1)
    return pd_frame

for year in years:
    for dirpath, dirnames, filenames in os.walk(os.path.join(euvs_dir,year)):
        # for filename in filenames:


            # file = os.path.join(dirpath,filename)

        files = glob.glob(os.path.join(dirpath,'*xrs*.csv'))
        for file in files:
            df = parseCSV2pandas(file)
            li.append(df)
        # try:
        #     # df = parseCSV2pandas(file)
        #     # pd.read_csv(file, index_col=None, header=0)
        #     # li.append(df)
        #     print(files)
        # except:pass

frame = pd.concat(li, axis=0, ignore_index=True)
frame.head()
```

```python


frame['time_tag'] = pd.to_datetime(frame['time_tag'])
# frame['B_FLUX'] = np.log(1 + frame['B_FLUX'] * 1e+9)
# frame['B_FLUX'] = (frame['B_FLUX'] * 1e+9) + 10
frame['B_FLUX_SCALED'] = np.log(frame['B_FLUX'].to_numpy() * 1e+9 + 1)

# frame.plot(x='time_tag',y='A_FLUX')
frame.head()
```
