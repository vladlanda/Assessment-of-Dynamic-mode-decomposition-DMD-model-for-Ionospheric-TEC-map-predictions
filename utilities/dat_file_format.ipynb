```python
import pandas as pd
import os
import glob
import re
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

% matplotlib qt
```

```python
euvs_dir = os.path.join(os.path.abspath(os.curdir),'..','EUVs')
euv_flag_files = glob.glob(os.path.join(euvs_dir,'*Flags*.dat'))
euv_dat_files = glob.glob(os.path.join(euvs_dir,'*Cnts*.dat'))
euv_dat_files,euv_flag_files
```

```python
# dat_file = euv_dat_files[0]

dat_df = pd.DataFrame()

for dat_file,flag_file in zip(euv_dat_files,euv_flag_files):
    with open(dat_file,'r') as file,open(flag_file,'r') as flag_file:
        year = re.findall(r'\d{4}',dat_file)[0]
        data_lines = file.readlines()[2:]
        flag_lines = flag_file.readlines()[2:]
        df_dict = {'datetime':[],'cnt':[],'flag':[],'year':[] } 
        for line,flag_line in zip(data_lines,flag_lines):

            d,t,c = line.split(" ")
            _,_,f = flag_line.split(" ")

            date = "{} {}".format(d,t)
            cnt = float(c.replace('\n',' '))
            flag = float(f.replace('\n',' '))

            df_dict['datetime'].append(date)
            df_dict['cnt'].append(cnt)
            df_dict['year'].append(year)
            df_dict['flag'].append(flag)

        dat_df = dat_df.append(pd.DataFrame(df_dict))

dat_df['datetime'] = pd.to_datetime(dat_df['datetime'])
# mask = (dat_df['flag'].shift(-3) != 0) | (dat_df['flag'].shift(3) != 0)
# mask = (dat_df['flag'].shift(-3) != 0) | (dat_df['flag'].shift(3) != 0)

# dat_df['cnt'].loc[mask] = np.nan
# dat_df['shift-5'] = dat_df['flag'].shift(-5)
# dat_df['shift+5'] = dat_df['flag'].shift(+5)
# mask = (dat_df['shift+5'] < 0) | (dat_df['shift+5'] > 0) | (dat_df['shift-5'] < 0) | (dat_df['shift-5'] > 0)


# dat_df['flag'].loc[mask] = np.nan
# dat_df['flag'] = dat_df['flag'] + 50000
# dat_df['cnt'].loc[mask] = np.nan
# dat_df['mask'] = [2000]*len(mask)
# dat_df['mask'].loc[mask] = 0

# dat_df['cnt_filtered'] = dat_df['cnt'].copy()
# mask = dat_df['flag'] > 0 #(dat_df['flag'] > 0) | (dat_df['flag'] < 0) | (dat_df['flag'] > 0) | (dat_df['flag'] < 0)
# dat_df['cnt_filtered'].loc[mask] = np.nan
# for s in range(0,15):
#     mask = mask | (dat_df['flag'].shift(-s) > 0) | (dat_df['flag'].shift(s) > 0)
#     pass
# for s in range(0,5):
#     mask = mask | (dat_df['flag'].shift(-s) < 0) | (dat_df['flag'].shift(s) < 0)
#     pass

# mask = mask | (dat_df['cnt_filtered'] < 50000)

# dat_df['cnt_filtered'].loc[mask] = np.nan

# dat_df['cnt_cp'] = dat_df['cnt'].copy()
# dat_df['cnt_cp'].loc[mask] = np.nan

dat_df
```
